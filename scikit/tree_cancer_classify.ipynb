## 유방암데이터를 이용해서 의사결정나무 만드는 코드

```python
from sklearn import datasets
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

def main():
  # load breast cancel data
  cancer = datasets.load_breast_cancer()

  x_train, x_test, y_train, y_test = train_test_split(cancer.data,
                                                      cancer.target,
                                                      stratify=cancer.target,
                                                      test_size=0.2,
                                                      random_state=42)
  # create cancer model
  cancer_model = DecisionTreeClassifier(criterion='entropy', max_depth=3)

  # learning model
  cancer_model.fit(x_train, y_train)
  print("Training Score: {:.3f}".format(cancer_model.score(x_train, y_train)))
  print("Testing Score: {:.3f}".format(cancer_model.score(x_test, y_test)))

  with open('cancer-dtree.dot', mode='w') as f:
    tree.export_graphviz(cancer_model, out_file=f,
                         feature_names=cancer.feature_names,
                         class_names=["cancer", "not cancer"])
if __name__ == '__main__':
  main()
```
